```python
'''Surging Pricing'''

def Price(distance,duration,base_fare,distance_unit,time_unit,minimum_fare):
    
    fare=base_fare+distance*distance_unit+time_unit*duration
    
    fare=max(fare,minimum_fare)
    
    return fare
```

```python
Order_df=pd.read_csv(os.path.join('Order_df_'+str(day)+'.csv'))

Order_df=Order_df.drop(columns=['Unnamed: 0'])

Order_df['Travel_duration']=Order_df.apply(lambda x:int(x['Travel_time']/6),axis=1)
    
Order_df['Travel_distance']=Order_df.apply(lambda x:(x['Travel_dis']/1000.0)*0.62,axis=1)

Order_df['Fare_A']=Order_df.apply(lambda x:Price(x['Travel_distance'],x['Travel_duration'],3,1.2,0.2,6),axis=1)

Order_df['Fare_B']=Order_df.apply(lambda x:Price(x['Travel_distance'],x['Travel_duration'],2.55,1.75,0.35,7),axis=1)

Order_df.to_csv(os.path.join('Order_df_'+str(day)+'.csv'))

Order_df
```
